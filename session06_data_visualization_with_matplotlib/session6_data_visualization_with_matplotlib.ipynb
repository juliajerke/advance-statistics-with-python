# 1. Introduction to Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib was develop by John Hunter in 2002 and 2003 and has been open source from the beginning on.

Website: https://matplotlib.org/stable/index.html

- Example plots
- Tutorials
- Documentation
- ...


```python
# Install the module via the console or the Anaconda navigator if it is not already installed

# Import the module
import matplotlib.pyplot as plt
```

## Components of a maplotlib figure

**Figure:**  
This is the entire figure object. Think of it as a canvas on which you draw your plot. The _figure_ keeps track of all the _axes_ that you include and all the different elements that you add to the _axes_.

**Axes:**  
This is the area in which you build and design your graph. A figure can contain more than one _axes_. An _axes_ contains two (2D case) or three _axis_ (3D case). Each _axes_ has elements such as title, x-label, y-label, etc.

**Axis:**  
Basically a number line. They set the limit of the graph and control the displayed ticks.

**"Artists":**  
Everything that you see in a figure is an _artist_. That is, _axes_, _axis_, text, lines, dots, etc. Most _artists_ are linked to a specific _axes_ and cannot be used in another _axes_.

# 2. Creating your first figures

#### 1 - Generating some data

Three variables that have been observed over a year in monthly intervals
- x
- y1
- y2

```python
# the numpy and random libraries will help us to create some fictional data
import numpy as np 
import random


# set seed for random module to ensure reproducibility
random.seed(a=53)

# creating or monthly time frame
month = range(1,13)
list(month)
```

Creating the variable x

```python
# x shall just contain some random numbers
x = [random.normalvariate(0,1)+2 for i in range(1,13)]  # list comprehension

# printing the first 5 numbers of x
x[:5] 
```

Creating the variable y1

```python
# creating a variable y1 that is dependent on x and illustrates a linear relationship
y1 = [i*1.7 + random.normalvariate(0,0.5) for i in x]
y1[:5]
```

Creating the variable y2

```python
# creating a second variable y2 that is also dependent on x but illustrates a quadratic relationship
y2 = [i**2 + random.normalvariate(0,0.5) for i in x]
y2[:5]
```

# 3. Data visualization with the World Happiness data set

#### 1 - Loading the data set

For that part we will be using the data from the world happiness report again.

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm

# Usually, you should be able to directly open the data set when it is in the same folder as the script
happy = pd.read_csv("world_happiness_2019.csv")
happy.head(20)
```

#### 2 - Data inspection

We will first inspect our data by 

- plotting a histogram of our variable of interest, the happiness score
- drawing a scatterplot with the happiness score and gdp

For the scatterplot we use the `pyplot.hist()` function

**Plotting the distribution of the happiness score**

We use the `axes.hist()` method to create a histogram. You can edit the number of bins, the range, the orientation, etc. See the documentation for more information: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html?highlight=hist#matplotlib.axes.Axes.hist

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

nbins = 15

# Adding the histogram to the plotting area
ax.hist("score",bins=nbins,density=False,data=happy,color="green") # changing density to False will plot frequencies instead of the density

ax.set_title("Distribution of the happiness score across the countries")
ax.set_xlabel("happiness")
ax.set_ylabel("Frequencies"); # Change the label to "Frequencies" if you change the density parameter from you scatter to False

```

**Scatterplot for bivariate or multivariate visualizations with continuous variables**

For the scatterplot we use the `pyplot.scatter()` function.

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

# Adding the scatter plot to the plotting area
ax.scatter(happy.gdp,happy.score,c="orange",marker="d");
```

Using the parameters `c` and `cmap`, we can color the points depending on a third variable and change the color pattern. See for an overview of available colormaps: https://matplotlib.org/stable/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py

The line `fig.colorbar(scatter,label="Social support")` adds a legend to the plot for the color shading of that additional variable. Note, that in this case you have to save the scatter plot into a variable so that you can call within the `colorbar` method.

The third variable is a categorical variable...

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

# Creating the scatter plot and adding a legend for the colormap
scat = ax.scatter(happy.gdp,happy.score,c=happy.develop,cmap="winter")

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scat.legend_elements(),title="Developing country")
ax.add_artist(legend1);
```

The third variable is a continuous variable...

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

# Creating the scatter plot and adding a legend for the colormap
sdhgls = ax.scatter(happy.gdp,happy.score,c=happy.support,cmap="autumn")
fig.colorbar(sdhgls,label="Social support");
```

#### 3 - Plotting regression results

We first have to run our regression and save the relevant information, such as the predicted values, the residuals, the lower and upper confidence bounds, etc., into new variables.

```python
# Describing the model
model = smf.ols("score ~ gdp", data = happy)

# Fitting the model
results = model.fit()

# Summarizing the model and printing a comprehensive overview
print(results.summary())
```

```python
# Saving the predicted values and the residuals to a new variable
happy["predicted"] = results.fittedvalues
happy["resid"] = results.resid

# Calculating the confidence intervals
prediction = results.get_prediction()
ci_lower = prediction.summary_frame()["mean_ci_lower"] # Access the lower value of the CI
ci_upper = prediction.summary_frame()["mean_ci_upper"] # Access the upper value of the CI

# Save the two values of the CI into a new data frame and append it to our existing data frame
conf_int = pd.DataFrame({"lower": ci_lower,"upper": ci_upper})
happy = happy.join(conf_int)

# Check whether everything worked as expected
happy.head(10)


### Note: If you get an error message for that cell, it is because you most likely run that cell more than once.
### In that case the join command will raise an error, because you are trying to join the variables for the confidence
### intervalls which already exist in the data set from a previous execution of that cell.
### The problem can be circumvented by saving the joined data set not into the happy data set but into a new one.
```

As before, we want to plot the happiness score against the gdp. But this time we also want to add the following to the graph:

- the predicted values using `axes.plot()`
- the upper and lower confidence bounds also using `axes.plot()`
- a confidence band, that is shading the area within the confidence bounds using `axes.fill_between()`

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

# Adding the scatter plot for the happiness score and the gdp
ax.scatter(happy.gdp, happy.score)

# Adding a line plot for the predicted happiness score
ax.plot(happy.gdp, happy.predicted,color="red")

# Adding the two lines for the lower and upper confidence interval
ax.plot(happy.gdp, happy.lower,color="gray",linewidth=0.5)
ax.plot(happy.gdp, happy.upper,color="gray",linewidth=0.5)

# Adding the confidence band
# the first value indicates the x coordinate, the other values indicate the coordinates of the lower and upper limit
# alpha is the transparency and ranges between 0 and 1
ax.fill_between(happy.gdp,happy.lower,happy.upper,color="gray",alpha=0.2)
```

You should see from the graph that something didn't quite work:

- the confidence band does not cover the entire range of our graph, and
- if you look close enough, that the lines for the confidence bounds are thicker in the mid-area

Remember the quotation from the beginning?

**"Matplotlib makes easy things easy and hard things possible."**

That might be one such case now. The "problem" is related to the logic of matplotlib. Matplotlib is mainly a tool for mathematical illustrations and graphs, and as such the standard data type that matplotlib understands is array data from numpy for instance. Here, in that case, we want to plot regression results based on a data frame.

This can be illustrated for the line  plot: A line plot adds dots to a graph and connects them by a line. But what dots will be connected?... Dots that are neighbors! Matplotlib determines the neighborhood of dots from the order of the input data - usually an array (in which the order usually plays a role). But in our case, their is no order in the data with respect to some form of _neighborhood_. Nevertheless, our data frame is ordered by the happiness score.
In the figure above, replace the line `ax.scatter(happy.gdp, happy.score)` with `ax.plot(happy.gdp, happy.score,"o")`. You will see that the look of the figure doesn't change. Now delete `"o"`. What do you see? The dots are now connected in ascending order by the happiness score - the way how our data is ordered!

To get a correct graph with confidence intervals and confidence band, we have to change the order of our data frame. We want to read the graph from left to right and, therefore, connect the dots from left to right. Hence, we have to order our data frame by the gdp instead of the happiness score. We will use `pd.sort_values()` for that.

```python
happy.sort_values(by=["gdp"],inplace=True)
happy.head(20)
```

We can now repeat the plot commands from before and will, eventually receive our expected output.

```python
# Initializing a figure object with one axes
fig, ax = plt.subplots()

# Adding the scatter plot for the happiness score and the gdp
ax.scatter(happy.gdp, happy.score)

# Adding a line plot for the predicted happiness score
ax.plot(happy.gdp, happy.predicted,color="red")

# Adding the two lines for the lower and upper confidence interval
ax.plot(happy.gdp, happy.lower,color="gray",linewidth=0.5)
ax.plot(happy.gdp, happy.upper,color="gray",linewidth=0.5)

# Adding the confidence band
# the first value indicates the x coordinate, the other values indicate the coordinates of the lower and upper limit
# alpha is the transparency and ranges between 0 and 1
ax.fill_between(happy.gdp,happy.lower,happy.upper,color="gray",alpha=0.2);
```

#### 4 - Regression plots from statsmodel

Relying on matplotlib, statsmodel offers the possibility to draw some specific regression plots. This includes for instance:

- prediction plots
- influence plots
- partial regression plots (in case of a multiple regression model)


See the website of statsmodel for more information: https://www.statsmodels.org/dev/examples/notebooks/generated/regression_plots.html

Estimating a new (this time multiple) regression model that includes the following variables:
- gdp per capita: _gdp_
- social support: _social_
- health life expectancy: _life_
- perceptions of corruption: _corruption_

```python
# Describing the model
model = smf.ols("score ~ gdp + support + life + corruption", data = happy)

# Fitting the model
results = model.fit()

# Summarizing the model and printing a comprehensive overview
print(results.summary())
```

**Fit plot**

```python
fig = sm.graphics.plot_fit(results, "gdp")
```

**Partial regression plots**

```python
fig = sm.graphics.plot_partregress_grid(results)
fig.tight_layout(pad=1.0)
```

**Regression diagnostics**

```python
fig = sm.graphics.plot_regress_exog(results, "gdp")
fig.tight_layout(pad=1.0)
```

**Influence plot**

```python
fig = sm.graphics.influence_plot(results, criterion="cooks")
fig.tight_layout(pad=1.0)
```
